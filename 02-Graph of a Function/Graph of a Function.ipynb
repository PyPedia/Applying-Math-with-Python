- To graph a function in Python using the SymPy library, we can use the plot function from sympy.plotting.
- SymPy is excellent for symbolic mathematics, and its plotting capabilities are simple and effective for visualizing functions.

**Step 2: Import Necessary Moduleslues using the function.**

```python
import sympy as sp
```

**Step 3: Define a Symbol**

Define the variable (symbol) for the function. This represents the independent variable:

```python
x = sp.symbols('x')
```

**Step 4: Define the Function**

Define the function you want to graph. For example:

```python
f = x**2 - 4  # A quadratic function
```

**Step 5: Plot the Function**

Use the *plot* function to graph it:

```python
sp.plot(f , (x, -5, 5))
```

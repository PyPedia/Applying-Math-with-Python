**1. What Are Polar Coordinates?**

Polar coordinates represent a point in a plane using:

- Radius (r): The distance of the point from the origin.
- Angle (θ): The angle measured from the positive x-axis to the line connecting the origin to the point.

A polar coordinate is written as (𝑟,𝜃).

**2. Polar and Cartesian Coordinates**

- In Cartesian coordinates, points are represented as (𝑥,𝑦).
- The relationship between polar and Cartesian coordinates:

$$ x=rcos(\theta) $$

$$ y=rsin(\theta) $$

$$ r=\sqrt{x^{2}+y^{2}} $$

$$\theta=tan^{-1}(\frac{y}{x})$$

**3. Mathematical Properties**

***Conversion Between Systems:***

- Polar to Cartesian:

$$ x=rcos(\theta) $$
$$ y=rsin(\theta) $$

- Cartesian to Polar:

$$ r=\sqrt{x^{2}+y^{2}} $$

$$\theta=tan^{-1}(\frac{y}{x})$$

- Area of a Sector:

The area 𝐴 of a sector in polar coordinates is:

$$A=\frac{1}{2}r^{2}d\theta$$

where Δ𝜃 is the angular span of the sector.

**4. Symmetry in Polar Coordinates**

- Symmetry about the x-axis: Replace 𝜃 with −𝜃.
- Symmetry about the y-axis: Replace 𝜃 with 𝜋−𝜃.
- Symmetry about the origin: Replace 𝑟 with −𝑟.

**5. Python Examples**

***Example 1: Converting Between Cartesian and Polar Coordinates***

```python
import math

# Cartesian to Polar
def cartesian_to_polar(x, y):
    r = math.sqrt(x**2 + y**2)
    theta = math.atan2(y, x)
    return r, math.degrees(theta)  # Returning theta in degrees for clarity

# Polar to Cartesian
def polar_to_cartesian(r, theta):
    theta_radians = math.radians(theta)  # Convert degrees to radians
    x = r * math.cos(theta_radians)
    y = r * math.sin(theta_radians)
    return x, y

# Example Usage
x, y = 3, 4
r, theta = cartesian_to_polar(x, y)
print(f"Cartesian ({x}, {y}) -> Polar ({r}, {theta} degrees)")

r, theta = 5, 53.13  # r and θ in degrees
x, y = polar_to_cartesian(r, theta)
print(f"Polar ({r}, {theta} degrees) -> Cartesian ({x}, {y})")
```

***Example 2: Plotting Polar Equations***

```python
import numpy as np
import matplotlib.pyplot as plt

# Plot a Rose Curve: r = a * cos(n * theta)
def plot_polar_equation(a, n):
    theta = np.linspace(0, 2 * np.pi, 1000)
    r = a * np.cos(n * theta)
    plt.polar(theta, r)
    plt.title(f"Rose Curve: r = {a} * cos({n} * θ)")
    plt.show()

plot_polar_equation(1, 5)
```

**6. Common Polar Equations**

***1. Circle***

- Equation: 𝑟=𝑎
- Represents a circle with radius 𝑎 centered at the origin.

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_circle(radius):
    theta = np.linspace(0, 2 * np.pi, 500)
    r = np.full_like(theta, radius)  # r is constant
    plt.polar(theta, r)
    plt.title(f"Circle: r = {radius}")
    plt.show()

plot_circle(5)
```

***2. Spiral of Archimedes***

- Equation: 𝑟=𝑎𝜃.
- As 𝜃 increases, 𝑟 grows linearly, forming a spiral.

```python
def plot_spiral(a):
    theta = np.linspace(0, 4 * np.pi, 1000)  # Cover multiple rotations
    r = a * theta
    plt.polar(theta, r)
    plt.title(f"Spiral: r = {a} * θ")
    plt.show()

plot_spiral(0.5)
```

***3. Rose Curves***

- Equation: 𝑟=𝑎cos(𝑛𝜃) or 𝑟=𝑎sin(𝑛𝜃)
- If 𝑛 is even, the curve has 2𝑛 petals.
- If 𝑛 is odd, the curve has 𝑛 petals.

```python
def plot_rose(a, n):
    theta = np.linspace(0, 2 * np.pi, 1000)
    r = a * np.cos(n * theta)
    plt.polar(theta, r)
    plt.title(f"Rose Curve: r = {a} * cos({n}θ)")
    plt.show()

plot_rose(2, 4)  # 4 petals (n is even)
plot_rose(3, 5)  # 5 petals (n is odd)
```

***4. Lemniscate***

- Equation:

$$r^{2}=a^{2}cos(2\theta)$$
or
$$r^{2}=a^{2}sin(2\theta)$$

- Represents a figure-eight curve.

```python
def plot_lemniscate(a):
    theta = np.linspace(0, 2 * np.pi, 1000)
    r_squared = a**2 * np.cos(2 * theta)
    r = np.sqrt(np.maximum(0, r_squared))  # Avoid imaginary values
    plt.polar(theta, r)
    plt.title(f"Lemniscate: r² = {a}² * cos(2θ)")
    plt.show()

plot_lemniscate(3)
```

***5. Cardioid***

- Equation: 𝑟=𝑎(1+cos(𝜃)) or 𝑟=𝑎(1+sin(𝜃))
- Heart-shaped curve.

```python
def plot_cardioid(a):
    theta = np.linspace(0, 2 * np.pi, 1000)
    r = a * (1 + np.cos(theta))
    plt.polar(theta, r)
    plt.title(f"Cardioid: r = {a}(1 + cos(θ))")
    plt.show()

plot_cardioid(2)
```

**7. Combining Multiple Polar Plots**

You can combine multiple polar plots to compare different equations.

***Example: Comparing a Circle and a Rose Curve***

```python
def plot_multiple_polar():
    theta = np.linspace(0, 2 * np.pi, 1000)

    # Circle
    r1 = np.full_like(theta, 3)
    plt.polar(theta, r1, label="Circle: r = 3")

    # Rose Curve
    r2 = 2 * np.cos(5 * theta)
    plt.polar(theta, r2, label="Rose: r = 2 * cos(5θ)")

    plt.legend(loc="upper right")
    plt.title("Comparison of Polar Plots")
    plt.show()

plot_multiple_polar()
```

**11. Polar Data Analysis**

***Example1: Visualizing Wind Directions***

- Wind data (speed and direction) can naturally be visualized in polar coordinates.

```python
import matplotlib.pyplot as plt
import numpy as np

def wind_polar_plot(wind_directions, wind_speeds):
    plt.figure(figsize=(8, 8))
    ax = plt.subplot(111, polar=True)

    # Convert wind directions to radians
    wind_directions_radians = np.radians(wind_directions)

    # Scatter plot
    scatter = ax.scatter(wind_directions_radians, wind_speeds, c=wind_speeds, cmap='viridis', alpha=0.75)
    plt.colorbar(scatter, label='Wind Speed (m/s)')

    plt.title("Wind Direction and Speed", va='bottom')
    plt.show()

# Example wind data
wind_directions = [0, 45, 90, 135, 180, 225, 270, 315]  # in degrees
wind_speeds = [10, 15, 20, 25, 15, 10, 5, 12]  # in m/s
wind_polar_plot(wind_directions, wind_speeds)
```

***Example 2: Fourier Transform in Polar Form***

- Polar coordinates are used in signal processing, like converting 2D Fourier transforms into radial and angular components for circular objects or patterns.

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_fourier_in_polar():
    # Generate a simple 2D wave pattern
    x = np.linspace(-10, 10, 200)
    y = np.linspace(-10, 10, 200)
    X, Y = np.meshgrid(x, y)
    Z = np.sin(2 * np.pi * np.sqrt(X**2 + Y**2))

    # Convert to polar coordinates
    R = np.sqrt(X**2 + Y**2)
    Theta = np.arctan2(Y, X)

    plt.figure(figsize=(10, 5))

    # Plot the wave in Cartesian coordinates
    plt.subplot(1, 2, 1)
    plt.imshow(Z, extent=(-10, 10, -10, 10), origin='lower', cmap='viridis')
    plt.title("Cartesian Wave Pattern")
    plt.colorbar()

    # Plot the radial component in polar
    plt.subplot(1, 2, 2, polar=True)
    plt.pcolormesh(Theta, R, Z, cmap='viridis')
    plt.title("Polar Wave Pattern")
    plt.colorbar()

    plt.tight_layout()
    plt.show()

plot_fourier_in_polar()
```
